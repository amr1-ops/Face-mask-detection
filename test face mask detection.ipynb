```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import models
```

```python
Classes = ["mask weared incorrect","with mask","without mask"]


model = models.load_model("Face mask Detection.model")
```

```python
def classifyImage(detectedFace):
    detectedFace = prepare_cnn(detectedFace)
    prediction = model.predict(np.array([detectedFace]))
    index=np.argmax(prediction)
    print("prediction = ",Classes[index])
    return Classes[index]
    
def prepare_cnn(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img=cv2.resize(img, (96, 96))
    plt.imshow(img,cmap=plt.cm.binary) 
    return img 
```

```python
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier('C:/Users/Ahmed Amr/anaconda3/pkgs/libopencv-4.0.1-hbb9e17c_0/Library/etc/haarcascades/haarcascade_frontalface_default.xml');



while True:
    ret, frame = cap.read()
    
    grayScaleFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(grayScaleFrame, 1.3, 5)
    
    for (x, y, w, h) in faces:
        
        faceX = x;
        faceY = y;
        faceWidth = w;
        faceHeight = h;
        detectedFace = frame[faceY:faceY+faceHeight, faceX:faceX+faceWidth] 
        classifyImage(detectedFace)
        cv2.putText(frame, classifyImage(detectedFace),(x, y - 10), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 200), 2)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 200), 4)
        
    cv2.imshow("Face mask detection", frame)
    key= cv2.waitKey(1)
    if key ==27:
        break

cap.release()
cv2.destroyAllWindows()
```
